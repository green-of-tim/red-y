```
from google.colab import files
import io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
uploaded = files.upload()
```

```
q_classic = pd.read_csv(io.BytesIO(uploaded['Q_tab_Taxi_30000_rewards.csv']))
q_double = pd.read_csv(io.BytesIO(uploaded['Q_double_tab_Taxi_30000_rewards.csv'])) 
```

```
plt.figure(figsize=(20,5))
plt.subplot(1, 2, 1)
plt.plot(q_classic['rewards'])
plt.plot(q_classic.rewards.rolling(100).mean())
plt.title('Classic Q-learning')
plt.ylabel('Rewards')

plt.subplot(1, 2, 2)
plt.plot(q_double['rewards'])
plt.plot(q_double.rewards.rolling(100).mean())
plt.title('Double Q-learning')
plt.ylabel('Rewards')
plt.figure(figsize=(20,10))
plt.show()
```

```
plt.figure(figsize=(20,5))

names=['Vanilla', 'Double', '5/30/500, d=0,618', '2/30/1000', '5/30/500, d=0,95', '10/30/100', '5/30/1000', '2/30/2000', '5/30/2000']
expected_reward=[5.6, 7.95, 7.2, 7.8, 7.9, 7.9, 8.061, 7.97, 7.9]
variance_reward=[436.3, 6.6, 86, 6.5, 6.5, 6.9, 6.3, 6.6, 6.69]
time=[11, 11.2, 44, 28.5, 43.2, 82, 64, 45, 109]

ax = plt.subplot(1, 3, 1)
plt.title('Average reward')
ax.bar(range(len(expected_reward)), expected_reward)
ax.set_xticks(range(len(expected_reward)))
ax.set_xticklabels(names, rotation=45)

ax = plt.subplot(1, 3, 2)
plt.title('Variance of reward')
ax.bar(range(len(variance_reward)), variance_reward)
ax.set_xticks(range(len(variance_reward)))
ax.set_xticklabels(names, rotation=45)

ax = plt.subplot(1, 3, 3)
plt.title('Learning time (seconds)')
ax.bar(range(len(time)), time)
ax.set_xticks(range(len(time)))
ax.set_xticklabels(names, rotation=45)

plt.show()
```
