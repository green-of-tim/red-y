```
from google.colab import files
import io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
uploaded = files.upload()
```

```
q_classic = pd.read_csv(io.BytesIO(uploaded['Q_tab_FL8x8_10000_steps.csv']))
q_double = pd.read_csv(io.BytesIO(uploaded['Q_double_tab_FL8x8_30000_steps.csv'])) 
```

```
plt.figure(figsize=(20,5))
plt.subplot(1, 2, 1)
plt.plot(q_classic['steps'])
plt.title('Classic Q-learning')
plt.ylabel('Steps')

plt.subplot(1, 2, 2)
plt.plot(q_double['steps'])
plt.title('Double Q-learning')
plt.ylabel('Steps')
plt.figure(figsize=(20,10))
plt.show()
```

```
plt.figure(figsize=(20,5))

names=['Vanilla', 'Double', '2/30/500', '5/30/500', '20/30/500', '100/30/500', '5/30/100', '2/30/1000']
expected_step=[83.3, 87.5, 86.3, 82.9, 89.7, 91, 110.8, 83.7]
variance_step=[1456, 1509, 1587, 1451.5, 1638.5, 1694, 1974, 1515]
time=[12.8, 67, 32.4, 79, 310, 0, 17.6, 70]

ax = plt.subplot(1, 3, 1)
plt.title('Average step')
ax.bar(range(len(expected_step)), expected_step)
ax.set_xticks(range(len(expected_step)))
ax.set_xticklabels(names, rotation=45)

ax = plt.subplot(1, 3, 2)
plt.title('Variance of step')
ax.bar(range(len(variance_step)), variance_step)
ax.set_xticks(range(len(variance_step)))
ax.set_xticklabels(names, rotation=45)

ax = plt.subplot(1, 3, 3)
plt.title('Learning time')
ax.bar(range(len(time)), time)
ax.set_xticks(range(len(time)))
ax.set_xticklabels(names, rotation=45)

plt.show()
```
